# Lets done some Exploratory Data Analysis on Univariate

## Catagorical column (Dytpe = Object)

```python
#Make countplot on Owner_type to show count the value of owner
sns.countplot(x = "Owner_Type",data =df)
```

```python
# Count the number of car brand 
sns.countplot(x='Brand', data=df, order=df['Brand'].value_counts().index)
plt.title('Count of Cars by Brand')
plt.xlabel('Brand')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
# Count the number of cars per fuel type
fuel_counts = df['Fuel_Type'].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(fuel_counts, labels=fuel_counts.index, autopct='%1.1f%%', startangle=120, colors=sns.color_palette('pastel', len(fuel_counts)))
plt.title('Distribution of Cars by Fuel Type')
plt.show()
```

## Numerical Columns (Dtype = Int,Float)

```python
# Hist to Show Car Power Distribution
sns.histplot(df['Power'], bins=30, kde=True)
plt.title('Distribution of Car Power')
plt.xlabel('Power (in bhp)')
plt.ylabel('Frequency')
plt.show()
```

```python
# BoxPlot to show Mileage
sns.boxplot(x=df['Mileage'])
plt.title('Show distribution of mileage')
plt.xlabel('Value')
plt.show()
```

```python
# hist plot to show Car price distribution 
sns.histplot(df['Price'], bins=30, kde=True)
plt.title('Distribution of Car Prices')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Count the number of seat people want
counts = df['Seats'].value_counts()
plt.figure(figsize=(8,8))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Pie Chart of Categorical Column')
plt.show()
```

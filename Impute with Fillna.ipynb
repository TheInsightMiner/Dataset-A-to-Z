```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("/Users/namansisodia/Downloads/cars.csv")
df.head()
```

```python
df.info()
```

```python
df.isnull().mean()*100
```

```python
# replace missing Value Kilometer and Mileage with pandas mean and median
# working on Kilometer driven mean
mean = df["Kilometers_Driven"].mean()
df["Kilo_mean"]= df["Kilometers_Driven"].fillna(mean)

# working on kiloemter driven median
median_ki = df["Kilometers_Driven"].median()
df["kilo_median"] = df["Kilometers_Driven"].fillna(median_ki)

# working on milege mean
mean_Mil = df["Mileage"].mean()
df["Mil_mean"]=df["Mileage"].fillna(mean_Mil)

# woking on mileage median 
median_mil = df["Mileage"].median()
df["Mil_median"] = df["Mileage"].fillna(median_mil)
```

```python
df.tail(10)
```

```python
# always check the effect you get while impute mean
fig = plt.figure()
ax = fig.add_subplot(111)

# original variable distribution
df['Kilometers_Driven'].plot(kind='kde', ax=ax)

# variable imputed with the median
df['kilo_median'].plot(kind='kde', ax=ax, color='red')

# variable imputed with the mean
df['Kilo_mean'].plot(kind='kde', ax=ax, color='green')

# add legends
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```

```python
# always check the effect you get while impute mean
fig = plt.figure()
ax = fig.add_subplot(111)

# original variable distribution
df['Mileage'].plot(kind='kde', ax=ax)

# variable imputed with the median
df['Mil_median'].plot(kind='kde', ax=ax, color='black')

# variable imputed with the mean
df['Mil_mean'].plot(kind='kde', ax=ax, color='yellow')

# add legends
lines, labels = ax.get_legend_handles_labels()
ax.legend(lines, labels, loc='best')
```
